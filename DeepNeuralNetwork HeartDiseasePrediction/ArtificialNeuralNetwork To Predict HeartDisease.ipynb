```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("heart.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
X=data.drop("target",axis=1)
Y=data["target"]
X.head()
```

# Splitting test and training data

```python
X_train, X_test, Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

# Building Neural Networks

```python
model=Sequential()
model.add(Dense(units=1500, activation='relu',input_dim=len(X_train.columns)))
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics='accuracy')
```

```python
model.fit(X_train,Y_train,epochs=200, batch_size=32)
```

# Accuracy Score

```python
#Train data accuracy
y_TrainAccuracy = model.predict(X_train)
y_TrainAccuracy = [0 if val < 0.5 else 1 for val in y_TrainAccuracy]
accuracy_score(Y_train, y_TrainAccuracy)
```

```python
#Test Data accuracy
y_TestAccuracy = model.predict(X_test)
y_TestAccuracy = [0 if val < 0.5 else 1 for val in y_TestAccuracy]
accuracy_score(Y_test, y_TestAccuracy)
```

```python
#As the test accuracy and train accuracy doesnt have have large difference, Our model is free from overfitting
```
